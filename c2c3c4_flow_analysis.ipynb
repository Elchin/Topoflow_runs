```python
# Download the CPCRW flow rate data from the BNZ LTER website and store it in the same folder
# Downloaded file should be identical to q_file name
q_file='cpcrw_flow_1969-2015.txt'
df= pd.read_csv(q_file)
flow=Series(df.ix[:,2]).values
name=Series(df.ix[:,0]).values
date=Series(df.ix[:,1]).values
n=len(flow)
print n
print Series(df.ix[0,:]).keys()
```

```python
def do_time_ser_exist(ws_name,ws_code,data,start,end,q):
    n=len(data);a=[]; b=[];
    for i in range(n):
        if ws_name[i]==ws_code:
            if data[i]==start:
                a=i
                print data[i],a
            if data[i]==end:
                b=i
                print data[i],b
    if a==[] or b==[]:
        print 'NO DATA'
    else:
        qq = q[a:b]
        qq = qq[~np.isnan(qq)]
        if qq.size:
            print 'min_discharge=',np.min(qq), ' max_discharge=',np.max(qq), ',over # of hours:',b-a
        else:
            print 'EMPTY ARRAY (CORRUPTED DATA)'        
```

```python
# calculting total number of measurements for C2,C3 and C4
j=0;k=0;m=0;c2_flow={};c3_flow={};c4_flow={}
for i in range(n):
    if name[i]=='C2':
        c2_flow[j]=flow[i]
        j=j+1
    elif name[i]=='C3':
        c3_flow[k]=flow[i]
        k=k+1
    else:
        c4_flow[m]=flow[i]
        m=m+1
print j,k,m
```
