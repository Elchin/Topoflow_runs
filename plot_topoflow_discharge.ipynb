```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from scipy.interpolate import interp1d
```

```python
# NOTE: looks like can not be used as is, and that is I shifted a bit
# I called the extracted from the file created Maria - raw 
# and Q_mes corresponds to the shifted data
def plot_discharge(Q_file, plt_title):
    Q_raw = genfromtxt('C2_flow_rate_2008.csv', delimiter=',')
    wk = np.loadtxt(Q_file, skiprows=2, unpack=False)
    time = wk[:,0]; Q_cal = wk[:,1]
    x = np.linspace(200, 2500, num=len(Q_raw), endpoint=True)
    Q_mes = interp1d(x, Q_raw)
    myplot=plt.plot(time,Q_cal,'r--',x,Q_mes(x),'b-'); 
    plt.xlabel('Time (min)')
    plt.ylabel('$Q_{C2}$ Discharge ($m^3/s$)')
    plt.title(plt_title)
    plt.setp(myplot, 'linewidth', 3.0)
    plt.legend(['cal', 'mes'], loc='best')
    plt.grid(); plt.show()
```

**TEST1:** This case uses driver data form examples/C2_basin/Test1 dropobox as is.   

```python
Q_file = '/Users/elchin/TopoFlow_Runs/Caribou/C2/Test1/Test2_0D-Q.txt' 
plot_discharge(Q_file, 'TEST1: Discharge at C2 vs. time')
```
